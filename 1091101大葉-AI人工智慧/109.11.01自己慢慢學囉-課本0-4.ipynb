```
import numpy as np #使用前先匯入numpy套件
x = np.zeros((3,5,2)) #建立一個形狀為(3,5,2)且完素皆為0的陣列
x.ndim #顯示軸數屬性,有3個軸
x.shape #顯示形狀屬性,為(3,5,2)
x.dtype #dtype('float64'),64位元的浮點數
```

```
x = np.zeros(5)
y = x.astype('uint8') #轉出為uint8並指定給y變數(不會改變原資料,稱為'轉出')
y.dtype #dtype('uint8') #已轉出為uint8
x.dtype dtype('float64') #原資料不受影響
```
