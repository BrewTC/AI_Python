```
#2.1constant()方法
import tensorflow as tf 
```

```
tf.constant(1) #創建一個張量
```

```
tf.constant(1.) #創建一個浮點型張量
```

```
tf.constant(2, dtype=tf.double) #創建的同時指定數據類型
```

```
tf.constant([[1.,2.,3.],[4.,5.,6.]]) #通過傳入一個list參數創建
```

```
#2.2convert_to_tensor()方法
import numpy as np
```

```
tf.convert_to_tensor(np.ones([2,3])) #轉換為張量
```

```
tf.convert_to_tensor([[2.,3.],[3.,4.]])
```

```
#2.3創建元素為指定值的tensor
#(1)zeros() ones()
a = tf.zeros([2,3,3])
a
```

```
b = tf.ones([2,3])
b
```

```
#(2)zeros_like ones_like
tf.zeros_like(b) #仿照b的shape創建一個全為0的tensor
```

```
tf.ones_like(a) #仿照a的shape創建一個全為1的tensor
```

```
#fill() 填入
tf.fill([2,3],5) #創建元素全為5,shape為[2,3]的向量
```

```
#2.4隨機初始化
#(1)tf.random.normal() 從指定正態分布中隨機取值
tf.random.normal([2,3], mean=1, stddev=1)
```

```
#truncated_normal() 從指定的截斷正態分布中隨機取值,但取值在兩個標準差範圍內
tf.random.truncated_normal([2, 3], mean=1, stddev=1)
```
