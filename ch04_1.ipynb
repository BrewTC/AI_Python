```
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical
sample=np.array([   
  [[0],[1]],
  [[1],[1]]                ,
  [ [1],[2] ],
  [ [2],[0] ]
]  )

label=np.array( [1,2,0,1]   )
sample.shape #(4,2,1) 4筆資料, 每筆兩個時序當輸入


label=to_categorical(label)
label
sample=to_categorical(sample)
sample.shape #4,2,3 4筆資料, 每筆兩個時序當輸入
model=Sequential()      
f1=tf.keras.layers.SimpleRNN(units=3, input_shape=(2,3),
  activation='softmax', dropout=0.0, 
  recurrent_dropout=0.0,go_backwards=False )
model.add(f1)
#model.summary()# m*d+d + d*d, m=3, d=3, total=21
#Call arguments:
#  inputs: A 3D tensor, with shape [batch, timesteps, feature]
model.compile(optimizer=tf.keras.optimizers.RMSprop(),
   loss=tf.keras.losses.CategoricalCrossentropy(),
   metrics=['acc'])
model.fit(x=sample, y=label, epochs=500)

```

```
result=model.predict_classes(sample)
result #([1, 2, 0, 1])
```

```
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical
sample=np.array([   
  [[0],[1]],
  [[1],[1]]                ,
  [ [1],[2] ],
  [ [2],[0] ]
]  )

label=np.array( [1,2,0,1]   )
sample.shape #(4,2,1) 4筆資料, 每筆兩個時序當輸入


label=to_categorical(label)
label
sample=to_categorical(sample)
sample.shape #4,2,3 4筆資料, 每筆兩個時序當輸入
model=Sequential()      
f1=tf.keras.layers.SimpleRNN(units=10, input_shape=(2,3),
  activation='relu', dropout=0.0, 
  recurrent_dropout=0.0,go_backwards=False )
model.add(f1)
f2=tf.keras.layers.Dense(3,activation='softmax')
model.add(f2)
#model.summary()# m*d+d + d*d, m=3, d=3, total=21
#Call arguments:
#  inputs: A 3D tensor, with shape [batch, timesteps, feature]
model.compile(optimizer=tf.keras.optimizers.RMSprop(),
   loss=tf.keras.losses.CategoricalCrossentropy(),
   metrics=['acc'])
model.fit(x=sample, y=label, epochs=500)

```

```
result=model.predict_classes(sample)
result #([1, 2, 0, 1])
```
